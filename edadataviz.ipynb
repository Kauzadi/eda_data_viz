First, let's try to see how the `FlightNumber` (indicating the continuous launch attempts.) and `Payload` variables would affect the launch outcome.

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

```python
### TASK  6: Visualize the launch success yearly trend
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract year from 'Date' column
df['Year'] = df['Date'].dt.year

# Calculate success rate for each year
yearly_success_rate = df.groupby('Year')['Class'].mean().reset_index()

# Plot success rate yearly trend
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Class', data=yearly_success_rate, marker='o', color='b')
plt.xlabel('Year')
plt.ylabel('Success Rate')
plt.title('Yearly Launch Success Rate Trend')
plt.grid(True)
plt.tight_layout()
plt.show()
```

You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Function to extract years from the date
def extract_year(date):
    return date.year

# Apply the function to extract years and create a new column 'Year'
df['Year'] = df['Date'].apply(extract_year)

# Calculate average success rate for each year
yearly_avg_success_rate = df.groupby('Year')['Class'].mean().reset_index()

# Plot average success rate yearly trend
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Class', data=yearly_avg_success_rate, marker='o', color='b')
plt.xlabel('Year')
plt.ylabel('Average Success Rate')
plt.title('Average Launch Success Rate Trend')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
## Features Engineering
import pandas as pd

# Function to extract year from the date
def extract_year(date):
    return pd.to_datetime(date).year

# Function to extract month from the date
def extract_month(date):
    return pd.to_datetime(date).month

# Function to calculate days since last launch
def days_since_last_launch(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.sort_values(by='Date')
    df['DaysSinceLastLaunch'] = df['Date'].diff().dt.days
    return df

# Function to calculate days until next launch
def days_until_next_launch(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.sort_values(by='Date')
    df['DaysUntilNextLaunch'] = df['Date'].diff(-1).dt.days
    return df

# Apply functions to create new features
df['Year'] = df['Date'].apply(extract_year)
df['Month'] = df['Date'].apply(extract_month)
df = days_since_last_launch(df)
df = days_until_next_launch(df)

# Display the updated dataframe with new features
print(df.head())
```

Use the function <code>get_dummies</code> and <code>features</code> dataframe to apply OneHotEncoder to the column <code>Orbits</code>, <code>LaunchSite</code>, <code>LandingPad</code>, and <code>Serial</code>. Assign the value to the variable <code>features_one_hot</code>, display the results using the method head. Your result dataframe must include all features including the encoded ones.


```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Define categorical columns
categorical_cols = ['Orbit_ES-L1', 'Orbit_GEO', 'Orbit_GTO', 'Orbit_HEO', 'Orbit_ISS', 'Orbit_LEO',
                    'Orbit_MEO', 'Orbit_PO', 'Orbit_SO', 'Orbit_SSO', 'Orbit_VLEO',
                    'LaunchSite_CCAFS SLC 40', 'LaunchSite_KSC LC 39A', 'LaunchSite_VAFB SLC 4E',
                    'LandingPad_5e9e3032383ecb267a34e7c7', 'LandingPad_5e9e3032383ecb554034e7c9',
                    'LandingPad_5e9e3032383ecb6bb234e7ca', 'LandingPad_5e9e3032383ecb761634e7cb',
                    'LandingPad_5e9e3033383ecbb9e534e7cc', 'Serial']

# Apply OneHotEncoder to categorical columns
encoder = OneHotEncoder(sparse=False)
features_one_hot = encoder.fit_transform(df[categorical_cols])

# Convert one-hot encoded features to DataFrame
features_one_hot_df = pd.DataFrame(features_one_hot, columns=encoder.get_feature_names_out(categorical_cols))

# Display the one-hot encoded features DataFrame
print(features_one_hot_df.head())

```
